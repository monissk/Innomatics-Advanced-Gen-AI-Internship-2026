```python
#1.Unique Words in a Sentence

sentence = "python is easy and python is powerful"

#Split sentence into words
words = sentence.split()

#Convert list to set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#2.Highest Salary

employees = {
    "Monis": 75000,
    "Ravi": 68000,
    "Kiran": 72000
}

#Find highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#3.Max and Min

numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
#4.Products above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
#5.Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

```python
#6.Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
#7.Character Frequency

text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Character Frequency:", frequency)
```

```python
#8.Convert List to Tuple

numbers_list = [10, 20, 30]

numbers_tuple = tuple(numbers_list)

print("Tuple:", numbers_tuple)
```

```python
#9.Check if Key Exists

employees = {
    "Monis": 75000,
    "Ravi": 68000
}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#10.Average Marks

marks = [70, 75, 65, 80]

total = sum(marks)
count = len(marks)

average = total / count

print("Average Marks:", average)
```
