```python
#1.Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    if amount not in valid_plans:
        print("Invalid plan selected.")
        return False
    
    print("Recharge Successful!")
    return True


#Retry system using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        if validate_recharge(amount):
            break
    except ValueError:
        print("Please enter a valid number.")
```

```python
#2.Inventory Reorder Alert System

def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(inventory)
```

```python
#3.Student Result Processing System

def process_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return f"Average: {average:.2f} → Pass"
    else:
        return f"Average: {average:.2f} → Fail"


student_marks = [60, 70, 45, 80]
print(process_result(student_marks))
```

```python
#4.Cab Fare Estimator

def calculate_fare(distance, peak_hour):
    base_fare = 50
    fare = base_fare + (12 * distance)
    
    if peak_hour:
        fare *= 1.25  # 25% extra
    
    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        
        peak_hour = True if peak_input == "yes" else False
        
        total_fare = calculate_fare(distance, peak_hour)
        print(f"Total Fare: ₹{total_fare:.2f}")
        break
        
    except ValueError:
        print("Invalid input. Please try again.")
```

```python
#5.Attendance Eligibility Checker

def check_attendance(attendance):
    present_count = 0
    
    for day in attendance:
        if day == "P":
            present_count += 1
    
    percentage = (present_count / len(attendance)) * 100
    
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"


attendance_record = ["P", "P", "A", "P", "P", "A", "P"]
print(check_attendance(attendance_record))
```

```python
#6.Password Strength Checker

def check_password(password):
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


user_password = input("Enter password: ")
print(check_password(user_password))
```
