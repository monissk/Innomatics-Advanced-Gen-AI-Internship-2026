```python
#1.Smart Parking Lot Management System

def manage_parking(capacity, logs):
    current_count = 0
    peak_usage = 0
    
    for log in logs:
        if log == "IN":
            current_count += 1
        elif log == "OUT" and current_count > 0:
            current_count -= 1
        
        # Track peak usage
        if current_count > peak_usage:
            peak_usage = current_count
    
    print("Currently Parked Vehicles:", current_count)
    
    if current_count > capacity:
        print("Parking Status: Exceeded Capacity!")
    else:
        print("Parking Status: Available")
    
    print("Peak Parking Usage:", peak_usage)


capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

```python
#2.Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  #5 minutes per km
    
    #Traffic impact
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 8
    
    #Weather impact
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20
    
    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")
```

```python
#3.Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats):
    occupied = len(booked_seats)
    occupancy_percentage = (occupied / total_seats) * 100
    
    print("Occupancy:", f"{occupancy_percentage:.0f}%")
    
    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    print("Show Status:", status)


total_seats = 200
booked_seats = [1] * 150  #150 seats booked

analyze_occupancy(total_seats, booked_seats)
```

```python
#4.Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0
    
    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    print("Average CPU Load:", f"{average:.0f}%")
    
    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    print("Server Status:", status)


cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

```python
#5.Smart Classroom Resource Usage Monitor

def monitor_resources(resource_usage):
    overused = []
    
    for resource, hours in resource_usage.items():
        if hours > 8:  #threshold for overuse
            overused.append(resource)
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resources)
```

```python
#6.Online Event Registration Capacity Controller

def manage_registration(capacity, registrations):
    
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    
    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"
    
    print("Registration Status:", status)


manage_registration(100, 105)
```
